```python
import pandas as pd

# Load the Excel file
file_path = '/content/DA - Task 1..xlsx'

# Show sheet names
excel_file = pd.ExcelFile(file_path)
excel_file.sheet_names
```

```python
#DA - Task 1.# Load data and taxonomy
df = pd.read_excel(file_path, sheet_name='Task')
taxonomy = pd.read_excel(file_path, sheet_name='Taxonomy')

df.head()        # show first few rows
taxonomy.head()  # show taxonomy list

```

```python
import re

def normalize(text):
    if pd.isna(text):
        return ""
    text = text.lower()
    text = re.sub(r'[^a-z0-9\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

for c in ['Complaint', 'Cause', 'Correction']:
    df[c + '_clean'] = df[c].apply(normalize)

df.head()
```

```python
mapping = {
    'loose': 'Loose Connection',
    'corrosion': 'Corrosion',
    'overheat': 'Overheating',
    'broken': 'Mechanical Damage',
    'short': 'Short Circuit',
    'connector': 'Connector Problem',
    'wire': 'Wiring Issue',
    'replaced': 'Component Replacement',
    'tightened': 'Tightening Required',
    'cleaned': 'Cleaning Required'
}
```

```python
def find_tags(text, mapping):
    for key, value in mapping.items():
        if key in text:
            return value
    return 'Unmapped'

df['Tag'] = df['Complaint_clean'].apply(lambda x: find_tags(x, mapping))
df.head(10)
```

```python
df.to_excel('AnishaK_AxionRay_Task1_tagged.xlsx', index=False)
```
