```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_path = '/content/DA -Task 2..xlsx'
df2 = pd.read_excel(file_path)
df2.head()
```

```python
df2.info()
df2.describe(include='all')
```

```python
summary = []
for col in df2.columns:
    summary.append({
        'Column': col,
        'Missing_Values': df2[col].isna().sum(),
        'Unique_Values': df2[col].nunique()
    })
pd.DataFrame(summary)
```

```python
# Fill missing numeric values with median
for col in df2.select_dtypes(include='number').columns:
    df2[col].fillna(df2[col].median(), inplace=True)

# Fill missing text with 'Unknown'
for col in df2.select_dtypes(include='object').columns:
    df2[col].fillna('Unknown', inplace=True)
```

```python
# 1. Bar chart of top 10 components (replace with real column name)
df2['GLOBAL_LABOR_CODE_DESCRIPTION'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Global Labor Code Descriptions')
plt.xlabel('Global Labor Code Description')
plt.ylabel('Count')
plt.show()

# 2. Histogram of numeric column
df2['REPAIR_AGE'].hist()
plt.title('Distribution of Repair Age')
plt.show()

# 3. Boxplot (replace with correct columns)
# Removed due to no clear 'Severity' column
# sns.boxplot(x='Severity', y='Repair_Time', data=df2)
# plt.title('Repair Time by Severity')
# plt.show()
```
